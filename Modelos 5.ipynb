### Revis√£o Te√≥rica

Uma s√©rie temporal √© uma sequ√™ncia de observa√ß√µes registradas em intervalos de tempo consistentes. Cada ponto da s√©rie representa um valor medido em um momento espec√≠fico, e, diferentemente de conjuntos de dados independentes, os valores de uma s√©rie temporal frequentemente dependem dos pontos anteriores, refletindo padr√µes como tend√™ncia, sazonalidade ou ciclos [4].

Existem diversos m√©todos para realizar previs√µes de s√©ries temporais. Dentre elas, Exponential Smoothing √© um m√©todo popular de previs√£o de s√©ries temporais, conhecido por sua simplicidade e precis√£o na previs√£o de tend√™ncias futuras com base em dados hist√≥ricos [5]. Esse m√©todo pressupoe que os padr√µes futuros ser√£o semelhantes aos dados passados recentes e se concentra em aprender o n√≠vel m√©dio de demanda ao longo do tempo.  O termo "suaviza√ß√£o exponencial" refere-se ao fato de que cada observa√ß√£o de demanda recebe um peso exponencialmente decrescente. [5] 

A previs√£o de suaviza√ß√£o exponencial pode ser dividida em tr√™s tipos principais: Suaviza√ß√£o Exponencial Simples ou √önica, Suaviza√ß√£o Exponencial Dupla e Suaviza√ß√£o Exponencial de Holt-Winters. Neste presente trabalho, ser√° utilizada a Suaviza√ß√£o Simples, m√©todo de previs√£o usado para dados de s√©ries temporais que n√£o apresentam tend√™ncia ou sazonalidade. [5]

Por fim, para an√°lise do modelo, foi utilizada a m√©trica Erro Percentual Absoluto M√©dio (MAPE), tamb√©m conhecido como Desvio Percentual Absoluto M√©dio (MAPD). Essa m√©trica mostra a diferen√ßa percentual m√©dia entre os valores reais e previstos [6]. Um MAPE menor significa melhor precis√£o de previs√£o e um MAPE de 0 significa previs√µes perfeitas.

* M (M√©dia): calcula a m√©dia de todas as observa√ß√µes
* A (Absoluto): usa valores absolutos para que erros positivos e negativos n√£o se cancelem
* P (Porcentagem): expressa o erro como uma porcentagem do valor real
* E (Erro): representa o qu√£o distantes as previs√µes est√£o dos valores reais

Durante a leitura, a coluna "date" foi convertida para o tipo `datetime`, garantindo que os dados fossem reconhecidos como s√©rie temporal. Essa convers√£o √© fundamental para permitir opera√ß√µes baseadas em tempo, como ordena√ß√£o cronol√≥gica, c√°lculo de diferen√ßas di√°rias e defini√ß√£o de frequ√™ncias regulares para modelagem de s√©ries temporais.

Os dados foram agrupados por data, somando casos e √≥bitos de todo o Brasil.

Para analisar a evolu√ß√£o da COVID-19 no Brasil, foi necess√°rio criar uma s√©rie temporal que representasse novos casos por dia, a partir dos casos acumulados. O procedimento foi realizado em etapas:

* **Defini√ß√£o do √≠ndice temporal**: a coluna "date" do DataFrame foi utilizada como √≠ndice da s√©rie `(set_index("date"))`, permitindo opera√ß√µes e an√°lises baseadas em datas.
  
* **C√°lculo dos novos casos di√°rios**: a fun√ß√£o de diferen√ßa di√°ria `(diff())` foi aplicada sobre a coluna de casos acumulados, gerando o n√∫mero de casos novos em cada dia. Essa transforma√ß√£o converteu os dados acumulativos em uma s√©rie de incid√™ncia di√°ria.

* **Tratamento de valores ausentes**: o primeiro valor da s√©rie resultante da diferen√ßa √© naturalmente NaN, pois n√£o h√° um dia anterior para subtrair. Este valor foi substitu√≠do por zero `(fillna(0))`, garantindo que a s√©rie estivesse completa desde o primeiro dia.

* **Defini√ß√£o de frequ√™ncia temporal**: para assegurar que o √≠ndice da s√©rie estivesse regular e com intervalos di√°rios consistentes, a frequ√™ncia foi explicitamente definida como di√°ria `(asfreq("D"))`. Essa etapa √© importante para m√©todos de previs√£o, que dependem de intervalos de tempo uniformes.

* **Filtragem de dias sem casos**: por fim, dias com zero casos foram removidos `(y[y > 0])` para focar apenas no per√≠odo de transmiss√£o ativa e evitar que per√≠odos sem incid√™ncia interferissem na modelagem.

  

A sintaxe da c√©lula a seguir foi baseada na refer√™ncia [3] gerada pelo Chat GPT [1]

```python
# Baixar dataset
path = kagglehub.dataset_download("unanimad/corona-virus-brazil")
print("Path:", path)

# Dentro dessa pasta, pegamos o arquivo certo (brazil_covid19.csv)
csv_path = os.path.join(path, "brazil_covid19.csv")

# Ler o CSV
df = pd.read_csv(csv_path, parse_dates=["date"])

# agrupar casos por data (Brasil inteiro)
df_brazil = df.groupby("date")[["cases", "deaths"]].sum().reset_index()

# S√©rie temporal: novos casos por dia
y = df_brazil.set_index("date")["cases"].diff().fillna(0)  # diferen√ßa = novos casos/dia
y = y.asfreq("D").fillna(0)

print(y.head())
```
